```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 12, 7]

titulo = "Scatterplot: Grafico de dispersao"
eixox = "Tempo (horas)"
eixoy = "Quantidade"

plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y, label = 'Individuos estudados', color = "#F18904", marker = "h", s = 60)
plt.plot(x, y, color = "#36688D") # comando linestyle muda o estilo da linha
plt.legend()

plt.show()
```
