### Primeiro teste de geração de Gráfico Boxplot com valores aleatórios.
#### Gerando 200 valores aleatórios que variam no intervalo de números 0 a 50.

```python
import matplotlib.pyplot as plt
import random

vector = []

for i in range(200):
    num_random = random.randint(0,50)
    vector.append(num_random)

plt.boxplot(vector)
plt.title("Boxplot de Valores Aleatórios")
plt.show()
```
