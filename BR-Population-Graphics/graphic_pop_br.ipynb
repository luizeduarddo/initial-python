### Gráfico de linhas e barras da evolução da população brasileiro entre os anos 1980 ao ano de 2022.
---

##### De acordo com os dados disponibilizados pelo Instituto Brasileiro de Geografia e Estatística (IBGE).

```python
import matplotlib.pyplot as plt

dados = open("populacao_br_atual.csv").readlines()

x = []
y = []

for i in range(len(dados)):
    if i != 0:
        linha = dados[i].split(";")
        x.append(int(linha[0]))
        y.append(int(linha[1]))
plt.plot(x, y, color = "#228B22", linestyle = "--")
plt.bar(x , y, color = "#DCDCDC")
plt.title("Crescimento da População Brasileira 1980-2022")
plt.xlabel("Ano")
plt.ylabel("População (100 milhões)")

plt.show()
```

##### Se deve levar em consideração que todos os dados da população brasileira foram fornecidos pelo IBGE, entretanto apenas os dados dos anos de 1980, 1991, 2000, 2010 e 2022 são dados confirmados com as pesquisas do Censo Demográfico do Brasil, realizado pelo IBGE, os dados dos outros anos são estimativas populacionais divulgadas pelo IBGE através de estudos estatísticos.
##### De acordo com o gráfico acima podemos ver a evolução constante da população brasileira no decorrer dos anos, tendo como o ano de 2022 o último ano a ter informação populacional. É possível observar algumas quedas sutis no gráfico entre os dados dos anos em que antecederam os anos dos Censos demográficos e os dados dos anos do Censo, devido as pesquisas de estimativa de população, provavelmente mostrarem a população um pouco mais elevada que a população real.
